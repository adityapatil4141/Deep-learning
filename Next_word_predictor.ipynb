```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs]) #We can send multiple texts at once, so input goes in a list
```

```python
tokenizer.word_index
```

```python
# Finding all the sentences in the dataset:
for sentence in faqs.split('\n'):
  print(sentence)
```

```python
# To convert text to numbers:
input_senquence = []
for sentence in faqs.split('\n'):
  token_Sent = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(token_Sent)):
    input_senquence.append(token_Sent[:1+i])
    # it will iterate for 3 times in first sentence -->[93,1,13], so for first iteration-->1,
    #it will be like token_sent[:(1+1)-->2]==>token_sent[:2]means all values till 2nd place means-->[93,1]
    #for next i==>3 it will be token_sent[:3], in our list index==> 93->0, 1->1, 13->12


# we use [sentence])[0], [0] because it was forming list inside a list to to select the first list we use [0]
#  According to output [93,1,13] corresponds to "About The Program" Which is indeed our first line
```

```python
input_senquence
```

```python
max_len = max([len(x) for x in input_senquence])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_input_sequences=pad_sequences(input_senquence,maxlen=max_len,padding='pre')
# we are doing padding to make all the sentence of similar lengeth. Every supervised learning model will work this way.
```

```python
padded_input_sequences
```

```python
x = padded_input_sequences[:,:-1] #We do not want last column meaning in first list we do not want '1' as it will be our output.prediction
y= padded_input_sequences[:,-1]
```

```python
x.shape,y.shape
```

```python
#one hot encoding
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=283) #num_classes is total.no of unique words in our input+1
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(283,100,input_length=56))
model.add(LSTM(150))
model.add(Dense(283,activation='softmax')) #nodes should be the number of words in the input
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x,y,epochs=100)
```

```python
import numpy as np
```

```python
text = 'The course follows'
for i in range(5):
  #tokeinze
  token_text = tokenizer.texts_to_sequences([text])[0]
  #padding
  padded_token_input_text=pad_sequences([token_text],maxlen=56,padding='pre')
  #prediction
  pos = np.argmax(model.predict(padded_token_input_text))  #we will get probability of each word from the input paragraph
  for word,index in tokenizer.word_index.items():
    if index==pos:
      text = text+" "+word
      print(text)
```
